```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

data = pd.read_csv("C:\\Users\\admin\\Downloads\\Dataset2.csv")
```

```python
# Count plot for distribution of settlements across provinces
plt.figure(figsize=(10, 6))
sns.countplot(x='Province', data=data)
plt.title('Distribution of Settlements Across Provinces')
plt.xlabel('Province')
plt.ylabel('Number of Settlements')
plt.xticks(rotation=45)
plt.show() 


```

```python
# Pie chart for distribution of establishment years
plt.figure(figsize=(8, 8))
data['Date of Establishment'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Distribution of Establishment Years')
plt.ylabel('')  # Remove y-label
plt.show()

```

```python
# Advanced charts
# 3. Scatter plot for correlation between initial population and current population
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Initial Population', y='Current Population', data=data)
plt.title('Correlation between Initial and Current Population')
plt.xlabel('Initial Population')
plt.ylabel('Current Population')
plt.show()
```

```python
# 4. Pairplot for exploring relationships between multiple variables
sns.pairplot(data, vars=['Initial Population', 'Current Population', 'Environmental Hazards', 'Socio-Economic Factors'])
plt.suptitle('Pairplot of Initial Population, Current Population, Environmental Hazards, and Socio-Economic Factors')
plt.show()
```

```python
# Boxplot for distribution of current population across provinces
plt.figure(figsize=(10, 6))
sns.boxplot(x='Province', y='Current Population', data=data)
plt.title('Distribution of Current Population Across Provinces')
plt.xlabel('Province')
plt.ylabel('Current Population')
plt.xticks(rotation=45)
plt.show()
```
